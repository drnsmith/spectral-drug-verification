# Spectral Drug Verification: Needle in a Haystack

This notebook explores a simplified drug verification problem using Raman spectral data.

The core challenge is a needle-in-a-haystack setting: two compounds may produce highly similar spectral fingerprints, and the task is to determine whether preprocessing, feature reduction, and modelling can improve detection of subtle differences while also supporting concentration verification.
